```python
def create_df(request):
    
    sneaker_features = ["id", "brand", "colorway", "gender", "name", "releaseDate", "retailPrice", "shoe", "styleId", "title", "year", "link"]
    final_df = pd.DataFrame(columns = sneaker_features)
    
    data = request['results']
    
    for shoe in range(0,len(data)):
        details = data[shoe]
        shoe_features = {}
        # Get metadata
        shoe_features["id"] = details["id"]
        shoe_features["brand"] = details["brand"]
        shoe_features["colorway"] = details["colorway"]
        shoe_features["gender"] = details["gender"]
        shoe_features["name"] = details["name"]
        shoe_features["releaseDate"] = details["releaseDate"]
        shoe_features["retailPrice"] = details["retailPrice"]
        shoe_features["shoe"] = details["shoe"]
        shoe_features["styleId"] = details["styleId"]
        shoe_features["title"] = details["title"]
        shoe_features["year"] = details["year"]
        shoe_features["link"] = details["media"]['smallImageUrl']
    
    
        # Concat the dfs
        shoe_df = pd.DataFrame(shoe_features, index = [0])
        final_df = pd.concat([final_df, shoe_df])
        
    return final_df
```

```python
create_df(dictionary)
```
